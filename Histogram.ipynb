```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#load an image
image_path= 'low_cont.jpg'
image= Image.open(image_path)
image
```

```python
#convert the image to grayscale if it is a color image
if image.mode != 'L':
    image= image.convert('L')
```

```python
#compute the histogram
histogram= image.histogram()
```

```python
plt.figure()
plt.title("Histogran of pixel Intensities")
plt.xlabel("Pixel Intesity")
plt.ylabel("Frequency")
plt.plot(histogram, color= "blue")
plt.show()
```

```python
image_paths= ['img1.jpg', 'img2.jpg', 'img3.jpg']
images= [Image.open(path) for path in image_paths]
```

```python
for i in range(len(images)):
    if images[i].mode != 'L':
        images[i]= images[i].convert('L')
```

```python
histograms3= [image.histogram() for image in images]
```

```python
plt.figure(figsize= (15,10))
for i in range(len(images)):
    plt.subplot(2,3, i+1)
    plt.imshow(images[i], cmap= 'gray')
    plt.title(f"Image {i+1}")
    plt.axis('off')
    
    plt.subplot(2,3, i+4)
    plt.plot(histograms3[i], color= 'blue')
    plt.title(f"Histogram {i+1}")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
    
    
```
